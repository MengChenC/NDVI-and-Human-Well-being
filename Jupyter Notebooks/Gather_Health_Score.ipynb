## Mental Health Score 

```python
df6 = pd.read_csv('VSRR_Provisional_Drug_Overdose_Death_Counts.csv')
df7 = pd.read_csv('suicide rate&death.csv')
```

```python
def select_df6_by_year(year):
    mask = (df6['Year'] == year)
    new_df6 = df6[mask]
    new_mask = (new_df6['Indicator']=="Number of Drug Overdose Deaths")
    new_df6 = new_df6[new_mask]
    
    return new_df6
```

```python
df18 = select_df6_by_year(2018)
```

```python
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```python
def drug_dose(lst,df):
    dic = {}
    for i in states:
        one_state = df[df["State"] == i]
        value1 = one_state['Predicted Value'].str.replace('[^\w\s]','')
        drug_rate = sum(value1.astype(int))
        dic[i] = drug_rate
    
    return dic
```

```python
drug18 = drug_dose(states,df18)
```

```python
new_drug18 = pd.DataFrame(list(drug18.items()),columns = ['State','Rate']) 
```

```python
drug_list18 = new_drug18.values.tolist()
```

```python
suicide = get_value_2018(df7)
suicide = suicide[["STATE","RATE"]]
suicide_list = suicide.values.tolist()
```

```python
#combine the 'Rate' from df6 and df7
dic = {}
for i in suicide_list:
    for j in drug_list18:
        if i[0] == j[0]:
            sum_ = i[1] + j[1]
        dic[i[0]] = sum_
```

```python
mental_results = pd.DataFrame(list(dic.items()),columns = ['State','Rate']) 
```

```python
dic_mental = mental_results.set_index('State').to_dict()
```

```python
lst2 = dic_mental['Rate']
lst2
```

```python
def sorted_rate(lst1,lst2):
    new_dic = {}
    for i in lst1.items():
        for j in lst2.items():
            if j[0] == i[0]:
                new_dic[i[0]] = j[1]
    return new_dic
```

```python
sorted_dic = sorted_rate(lst18,lst2) 
sorted_dic
```

```python
sorted_mental = pd.DataFrame(sorted_dic.items(), columns=['State', 'Sum'])
sorted_mental
```
